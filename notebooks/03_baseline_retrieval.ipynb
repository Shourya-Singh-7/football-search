```python
import sys
import os

PROJECT_ROOT = os.getcwd()
sys.path.append(PROJECT_ROOT)

from src.retrieve import SemanticRetriever

r = SemanticRetriever(
    index_path=os.path.join(PROJECT_ROOT, 'data/processed/baseline.index'),
    embeddings_path=os.path.join(PROJECT_ROOT, 'data/processed/baseline_embeddings.npy'),
    questions_path=os.path.join(PROJECT_ROOT, 'data/processed/questions.csv'),
    model_name='all-mpnet-base-v2',
)
print("Retriever loaded.")
```

```python
examples = []

test_queries = [
    "What is a false 9?",
    "How does the offside rule work?",
    "Why do teams play with three at the back?",
    "What's the difference between a regista and a deep-lying playmaker?",
    "How does VAR decide on penalties?",
    "What is gegenpressing?",
    "How does a football manager rotate their squad?",
    "Why is tiki-taka effective?",
    "What does it mean to press high up the pitch?",
    "How do wingers differ from inside forwards?",
    "Is a backline with 3 centre backs in a 3ATB the same as a backline with two wing backs in a 5ATB?",
    "What is the difference between a sweeper keeper and a ball playing keeper?",
    "What is the best way to break open a low-block?",
    "How do you best defend a high pressing system?",
    "How do you best defend against gegenpressing?",
    "What is the difference between a holding midfielder and a cdm?",
]

for query in test_queries:
    results = r.search(query, top_k=5)
    top1 = results[0] if results else None
    examples.append({
        'query': query,
        'top1_title': top1['title'] if top1 else '',
        'top1_score': round(top1['score'], 3) if top1 else 0,
        'top1_tags': top1['tags'] if top1 else '',
        'verdict': '',
        'failure_mode': '',
        'notes': '',
    })

print(f"Done. {len(examples)} queries logged.")
```

```python
import pandas as pd

log = pd.DataFrame(examples)
os.makedirs('results', exist_ok=True)
log.to_csv('results/error_analysis_baseline.csv', index=False)
log
```
