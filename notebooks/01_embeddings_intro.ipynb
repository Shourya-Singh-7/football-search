```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')  # small + fast for exploration
```

```python
# Encode some sentences
sentences = [
    "What is the el clasico?",
    "What is barcelona vs real madrid?",
    "What is the match between barcelona and real madrid?",
]
embeddings = model.encode(sentences)
print(embeddings.shape)  # (5, 384)
```

```python
# Compute similarity between all pairs
sim = cosine_similarity(embeddings)
print(sim)  # football sentences should be more similar to each other
```
